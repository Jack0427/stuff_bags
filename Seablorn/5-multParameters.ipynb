```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid',color_codes=True)

np.random.seed(sum(map(ord,'categorical')))
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.stripplot(x='day', y='total_bill',data=tips, jitter=True) # jitter 預設為true 降低重疊現象
```

```python
sns.swarmplot(x='day', y='total_bill',data=tips) # 較平均的jitter
```

```python
sns.swarmplot(x='day', y='total_bill',data=tips, hue='sex')
```

```python
sns.swarmplot(y='day', x='total_bill',data=tips, hue='time')
```

# 盒圖
* IQR 四分位距 Q3 - Q1
* N = 1.5IQE 如果一個值 > Q3+N or < Q1-N 則為離群點

```python
sns.boxplot(x='day', y='total_bill', hue='time',data=tips)
```

```python
sns.violinplot(x='total_bill', y='day', hue='time',data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex',data=tips,split=True) # 畫在一起
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,inner=None) 
sns.swarmplot(x='day', y='total_bill', data=tips,color='w',alpha=.5) 
```

```python
# 顯示值得集中趨勢
sns.barplot(x='sex', y='survived', data=titanic, hue='class') 
```

```python
# 點圖可以更好的描述變化差距
sns.pointplot(x='sex', y='survived', data=titanic, hue='class') 
```

```python
sns.pointplot(x='class', y='survived', data=titanic, hue='sex',palette={'male':'b','female':'m'},markers=['*','o'],linestyles=['-','--']) 
```

```python
# 寬行數據
sns.boxplot(data=iris,orient='h')
```

```python
# 多層面版分類圖
sns.factorplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', hue='smoker', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', hue='smoker', data=tips, kind='swarm', col='time')
```

```python
sns.factorplot(x='time', y='total_bill', hue='smoker', data=tips, kind='box', col='day',size=4, aspect=.5)
```
