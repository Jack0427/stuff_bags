```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(6,6)})
```

### 調色板
color_palette() 傳入matplotlib所支持的顏色
set_palette() 設置所有圖的顏色

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
sns.palplot(sns.color_palette('hls',20)) # 更多默認方案
```

```python
data = np.random.normal(size=(20,20)) + np.arange(20) / 2
sns.boxplot(data=data,palette=sns.color_palette('hls',20))
```

# xkcd
包含一套針對隨機RGB色的命名 產生954個可以通過xkcd_rgb調用的顏色名字

```python
plt.plot([0,1],[0,1],sns.xkcd_rgb['pale red'],lw=3)
plt.plot([0,1],[0,2],sns.xkcd_rgb['medium green'],lw=3)
plt.plot([0,1],[0,3],sns.xkcd_rgb['denim blue'],lw=3)
```

```python
colors = ['windows blue','amber','greyish','faded green','dusty purple']
sns.palplot(sns.xkcd_palette(colors))
```

```python
x, y = np.random.multivariate_normal([0,0],[[1,-5],[-.5,1]],size=300).T
pal = sns.dark_palette('green',as_cmap=True)
sns.kdeplot(x,y,cmap=pal)
```
