```python
for dtype in ['float64','int64','object']:
    selected = df.select_dtypes(include=[dtype])
    mean_usage_b = selected.memory_usage(deep=True).mean()
    mean_usage_mb = mean_usage_b / 1024**2 
    print(f'平均記憶體占用-{dtype}: {mean_usage_mb} MB')
```

```python
int_types = ['uint8','int8','int16','int32','int64',]
for it in int_types:
  print(np.iinfo(it)) # show max num
```

```python
def mean_usage(pandas_obj):
    if isinstance(pandas_obj,pd.DataFrame):
      usage_b = pandas_obj.memory_usage(deep=True).sum()
    else:
      usage_b = pandas_obj.memory_usage(deep=True)
    usage_mb = usage_b / 1024**2
    return '{:.2f} MB'.format(usage_mb)
df_int = df.select_dtypes(include=['int64']) # 要搜尋的資料型態 int float
coverted_int = df_int.apply(pd.to_numeric,downcast='unsigned')
print(mean_usage(df_int),mean_usage(coverted_int)) #降低記憶體使用
```

```python
converted_obj = pd.DataFrame()
df_obj = df.select_dtypes(include=['object'])
for col in df_obj.columns:
    num_unique_values = len(df_obj[col].unique())
    num_total_values = len(df_obj[col])
    if num_unique_values/num_total_values < 0.5: # 單一較少重複較多
        converted_obj.loc[:,col] = df_obj[col].astype('category')
    else:
        converted_obj.loc[:,col] = df_obj[col]
```

```python
print(mean_usage(df_obj))
print(mean_usage(converted_obj))
```
