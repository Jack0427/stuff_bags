```python
import math
x = np.random.normal(loc=0.0,scale=1.0,size=300)
width = 0.5
bins = np.arange(math.floor(x.min())-width,math.ceil(x.max())+width,width)
ax = plt.subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(right=True) #  bottom right left top
plt.grid()
plt.hist(x,alpha=0.5,bins=bins)
```

```python
x = np.arange(10)
for i in range(1,4):
  plt.plot(x,i*x**2,label=f'group {i}')
plt.legend(loc='center left')
```

```python
fig = plt.figure()
ax = plt.subplot(111)
x = np.arange(10)
for i in range(1,4):
  plt.plot(x,i*x**2,label=f'group {i}')
ax.legend(loc='upper center',bbox_to_anchor=(0.5,1.2),ncol=2) # ncol 
```

```python
fig = plt.figure()
ax = plt.subplot(111)
x = np.arange(10)
for i in range(1,4):
  plt.plot(x,i*x**2,label=f'group {i}')
ax.legend(loc='upper center',bbox_to_anchor=(1.15,.65),ncol=1)
```

```python
x = np.arange(10)
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = '20' # modify global attr
for i in range(1,4):
  plt.plot(x,i*x**2,label=f'group {i}',marker='o')
plt.title('sdfsdf')
plt.legend(loc='upper right',framealpha=.05)
```
