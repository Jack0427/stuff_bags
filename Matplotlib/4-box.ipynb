```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# value < Q1 - 1.5*IQR or Q3 + 1.5*IQR > value 離群值(outliers)
#
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
plt.boxplot(data,notch=False,sym='s',vert=True) # notch 特殊形狀 sym點的形狀 vert 垂直

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
bplot = plt.boxplot(data,notch=False,sym='s',vert=True) # notch 特殊形狀 sym點的形狀 vert 垂直

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')

for componts in bplot.keys():
    for line in bplot[componts]:
        line.set_color('c')
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
bplot = plt.boxplot(data,notch=True,sym='s',vert=False) # notch 特殊形狀 sym點的形狀 vert 垂直

plt.yticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
bplot = plt.boxplot(data,notch=False,sym='s',vert=True, patch_artist=True) 

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')

colors = ['pink','lightblue','lightgreen']

for pathch, color in zip(bplot['boxes'],colors):
    pathch.set_facecolor(color)
```

```python
# 小提琴圖 violinplot
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,5))
data = [np.random.normal(0,std,100) for std in range(6,10)]
axes[0].violinplot(data,showmeans=False,showmedians=True) # 越胖的地方 值越多
axes[0].set_title('violin plot')
axes[1].boxplot(data)
axes[1].set_title('box plot')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(data))])
plt.setp(axes,xticks=[y+1 for y in range(len(data))],xticklabels=['x1','x2','x3','x4'],)
```
