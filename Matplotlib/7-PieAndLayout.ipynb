```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 51212.
f = 40742.
m_perc = m/(m+f)
f_perc = f/(m+f)

colors = ['navy','lightcoral']
labels = ['Male','Female']

plt.figure(figsize=(5,5))
paches,texts,autotexts = plt.pie([m_perc,f_perc],labels=labels,autopct='%1.1f%%',explode=[0,0.05],colors=colors) # explod控制縫隙大小
for text in texts+autotexts:
    text.set_fontsize(20)
for text in autotexts:
  text.set_color('white') 
```

```python
x = np.linspace(0,10,10000)
y2 = np.sin(x**2)
y1 = x**2
fig, ax1 = plt.subplots()
left, bottom, width, height = [.2,.4,.3,.35]
ax2 = fig.add_axes([left, bottom, width, height])
ax1.plot(x,y1)
ax2.plot(x,y2)
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

def product(name):
    if name:
      return [name + str(i) for i in range(10)]
    else:
      return [round(i,2) for i in np.random.rand(10)]
def autlabel(rects):
  for rect in rects:
    hieght = rect.get_height()
    ax.text(rect.get_x()+rect.get_width()/2,1.02*hieght,'{:,}'.format(float(hieght)),ha='center',va='bottom',fontsize=18)
```
