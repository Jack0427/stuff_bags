```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = np.random.normal(0,20,1000)
bins = np.arange(-100,100,5)
plt.hist(data,bins=bins)
plt.xlim([min(data)-5,max(data)+5])
```

```python
import random
d1 = [random.gauss(15,10) for i in range(500)]
d2 = [random.gauss(5,5) for i in range(500)]
bins = np.arange(-50,50,2.5)

plt.hist(d1,bins=bins,label='class1',alpha=.3)
plt.hist(d2,bins=bins,label='class2',alpha=.3)
plt.legend()
```

```python
mu_vecl = np.array([0,0])
cov_matl = np.array([[2,0],[0,2]])

x1_samples = np.random.multivariate_normal(mu_vecl,cov_matl,100)
x2_samples = np.random.multivariate_normal(mu_vecl+.2,cov_matl+.2,100)
x3_samples = np.random.multivariate_normal(mu_vecl+.4,cov_matl+.4,100)

plt.figure(figsize=(10,8))
# print(x1_samples)
plt.scatter(x1_samples[:,0],x1_samples[:,1],marker='x',color='b',alpha=.6,label='x1') # 第一維數據
plt.scatter(x2_samples[:,0],x2_samples[:,1],marker='o',color='r',alpha=.6,label='x2')
plt.scatter(x3_samples[:,0],x3_samples[:,1],marker='*',color='g',alpha=.6,label='x3')
plt.legend()
```

```python

x_coords = [.25,.35,.95,.1,.5]
y_coords = [.25,.35,.95,.1,.5]
plt.figure(figsize=(10,8))
plt.scatter(x_coords,y_coords,marker='s',s=50) # s = size
for x,y in zip(x_coords,y_coords):
  plt.annotate(f'({x},{y})',xy=(x,y),xytext=(0,-15),textcoords = 'offset points',ha='center')
```

```python
mu_vecl = np.array([0,0])
cov_matl = np.array([[2,0],[0,2]])
X = np.random.multivariate_normal(mu_vecl,cov_matl,500)
plt.figure(figsize=(10,8))
R = X**2
R_sum = R.sum(axis=1)
plt.scatter(X[:,0],X[:,1],color='grey',marker='o',s=20*R_sum,alpha=.5)
```
